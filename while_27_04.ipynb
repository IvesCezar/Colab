#EXPLICAÇÃO


```python
#contagem regressiva
contador = 10
while contador > 0:
  print("Contador:", contador)
  contador-=1
```

```python
#Estrutura de Loop while
soma = 0
numero = 0

while numero != -1:
  numero = int(input("Digite um número (-1 para parar): "))
  if numero != -1:
      soma += numero
print("A soma dos números é:", soma)
```

```python
#Estrutura de Loop while
senha_correta = "1234"
tentativa = ""
while tentativa != senha_correta:
    tentativa = input ("Digite a senha:") and input ("Senha incorreta:")
print ("Acesso permitido!")
```
