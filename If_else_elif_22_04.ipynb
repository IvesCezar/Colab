```python
#Atividade 1
numero = float(input("Digite um numero:"))
if numero > 0:
  print("este numero é positivo")
elif numero == 0:
  print("numero nulo")
else:
  print("este numero é negativo")
```

```python
#Atividade 2
numero = float(input("digite um numero"))
if numero % 2 == 0:
  print("este numero é par")
else:
  print("este numero é impar")
```

```python
#Atividade 3
nota = float(input("digite a sua nota:"))
if nota >= 7:
  print("aprovado")
elif nota >= 5:
  print("recuperação")
else:
  print("reprovado")
```
