```python
#ATIVIDADE 1
p1 = int(input("Digite a sua idade:"))
p2 = int(input("Digite a sua idade:"))
print(p1 == p2)
print(p1 != p2)
print(p1 > p2)
print(p1 >= p2)
print(p1 < p2)
print(p1 <= p2)
```
