```python
# soma/sub/mult/div
valor1 = int(input("Digite um numero qualquer:"))
valor2 = int(input("Digite outro numero qualquer:"))
resultado1 = valor1 + valor2
resultado2 = valor1 - valor2
resultado3 = valor1 * valor2
resultado4 = valor1 / valor2

print ("Seu resultado é:", resultado1)
print ("Seu resultado é:", resultado2)
print ("Seu resultado é:", resultado3)
print (f"Seu resultado é: {resultado4:.0f}")
```
