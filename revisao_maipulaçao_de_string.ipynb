```python
#1
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "
nome_limpo = nome.strip()
email_minusculo = email.strip().lower()
email_novo = email_minusculo.replace("@empresa.com", "@gmail.com")
primeira_letra = nome_limpo[0]
ultima_letra = nome_limpo[-1]
print("Nome limpo:", nome_limpo)
print("Email final:", email_novo)
print("Primeira letra:", primeira_letra)
print("Última letra:", ultima_letra)
```
