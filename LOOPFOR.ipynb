```python
#Atividade 1
vendas_funcionarios = [2800, 3200, 1500, 4000, 5000, 1000, 3500]
meta_vendas = 3000
percentual_bonus = 0.05

for vendas in vendas_funcionarios:
    if vendas >= meta_vendas:
        bonus = vendas * percentual_bonus
        print(f"Vendas: R${vendas:.2f} - Bônus: R${bonus:.2f}")
    else:
        bonus = 0.00
        print(f"Vendas: R${vendas:.2f} - Bônus: R${bonus:.2f} (Meta não atingida)")
```

```python
# 2
vendas_diarias = [2000, 3000, 2500, 4000, 1800, 5000, 2700]
meta_vendas_diarias = 2500

dias_acima_da_meta = 0

for vendas in vendas_diarias:
    if vendas > meta_vendas_diarias:
        dias_acima_da_meta += 1

print(f"Número de dias com vendas acima da meta de R${meta_vendas_diarias:.2f}: {dias_acima_da_meta}")
```

```python
#4
vendas_vendedores = [1500, 2500, 1800, 3200, 4000, 1200, 2800]
meta_comissao = 2000
percentual_comissao = 0.08

for vendas in vendas_vendedores:
    if vendas > meta_comissao:
        comissao = vendas * percentual_comissao
        print(f"Vendas: R${vendas:.2f} - Comissão: R${comissao:.2f}")
    else:
        comissao = 0.00
        print(f"Vendas: R${vendas:.2f} - Comissão: R${comissao:.2f} (Meta não atingida)")
```

```python
#5
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_bonus = 0.1
bonus_extra = 50

for vendas in lista_vendas:
    bonus_total = 0
    if vendas >= meta:
        bonus_total = vendas * percentual_bonus
    if vendas > 2000:
        bonus_total += bonus_extra

    print(f"Venda: R${vendas:.2f} - Bônus Final: R${bonus_total:.2f}")
```

```python
#6
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

maior_venda = lista_vendas[0]

for venda in lista_vendas:
    if venda > maior_venda:
        maior_venda = venda

print(f"O maior valor de venda é: R${maior_venda:.2f}")
```
