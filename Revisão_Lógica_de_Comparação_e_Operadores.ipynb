```python
#1
idade1 = int(input("Digite a primeira idade: "))
idade2 = int(input("Digite a segunda idade: "))
print("Igual?", idade1 == idade2)
print("Diferente?", idade1 != idade2)
print("Maior?", idade1 > idade2)
```

```python
#2
idade = int(input("Digite sua idade: "))
possui_convite = input("Você possui convite? (sim/nao): ")
acesso = idade >= 18 and possui_convite == "sim"
print("Pode entrar?", acesso)
```

```python
#3
usuario_correto = "admin"
senha_correta = "1234"
usuario_digitado = input("Digite o usuário: ")
senha_digitada = input("Digite a senha: ")
login_valido = usuario_digitado == usuario_correto and senha_digitada == senha_correta
print("Login válido?", login_valido)
```
